```python
def zeroPad(img, h, w, h_pad, w_pad):
    # zero matrix
    ni_h = h + h_pad
    ni_w = w + w_pad
    new_img = np.zeros((ni_h, ni_w), 'uint8')
    
    # copy image into matrix
    n = 0
    for i in range(h_pad // 2, ni_h - h_pad // 2):
        m = 0
        for j in range(w_pad // 2, ni_w - w_pad // 2):
            new_img[i][j] = img[n][m]
            m = m + 1
        n = n + 1
    
    return new_img
```

```python
def flipFilter(f):
    return f[::-1, ::-1]
```

```python
def getPixelValue(padded_img, filter, i, j):
    f_h, f_w = filter.shape
    summed = 0
    for k in range(f_h):
        n_j = j
        for l in range(f_w):
            summed = summed + filter[k][l] * padded_img[i][n_j]
            n_j = n_j + 1
        i = i + 1
    return summed
```

### Linear filtering: correlation

```python
import numpy as np
```

```python
filter_gaussian = np.array([[1, 16, 1], [2, 4, 2], [1, 2, 1]], dtype = np.float64)
```

```python
filter_gaussian = filter_gaussian / 16
```

```python
filter_gaussian
```

```python
from pylab import *
```

```python
image = imread("image.jpg")
img_grey = (image[..., 0] + image[..., 1] + image[..., 2]) // 3
```

```python
x, y = img_grey.shape
img_after_filter = np.empty([508, 510])
```

```python
for i in range(1, 509):
    for j in range(1, 511):
        sum = 0
        for k in range(-1, 1):
            for p in range(-1, 1):
                sum = sum + img_grey[i+k][j+p] * filter_gaussian[k+1][p+1]
        img_after_filter[i-1][j-1] = sum
```

```python
np.sum(img_grey[5:8, 8:11] * filter_gaussian)
```

```python
for i in range(1, 509):
    for j in range(1, 511):
        img_after_filter[i-1][j-1] = sum(img_grey[i-1:i+2,j-1:j+2] * filter_gaussian)
```

```python
from PIL import Image
```

```python
img = Image.fromarray(img_after_filter)
img.show()
```
